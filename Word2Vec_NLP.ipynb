```python
from gensim.models import Word2Vec
```

```python
from nltk.corpus import stopwords
```

```python
import re
```

```python
paragraph = """My best regards to everyone present. To start with I would welcome you all and also thank you for giving me this marvelous opportunity. Dr. APJ Abdul Kalam is not just a name but a feeling. He was the 11th President of India. He was one of the greatest scientists and is known for his achievements in the field of science. He is also popularly known as the Missile man of India. He was a man of power, status, knowledge, and virtue. Born on 15 October 1931 in Rameswaram, Tamil Nadu, Dr. Kalam is a great motivator for many. He was born in a middle-class Muslim family and witnessed many struggles. Even after that he never gave up and moved ahead on the path of determination and success. His full name was Avul Pakir Jainulabdeen Abdul Kalam. Since the beginning, he had a very keen interest to study. He served DRDO and ISRO as the scientist. He was a man of dreams and ideas. He dreamt of making India one of the super-powers in the world. His idea of dreaming was really different. He emphasized that the dreams are not those which you see when you sleep but are those which never let you sleep. Some countless efforts and contributions are made by Dr. Kalam for the sake of the nation. He was awarded by Bharat Ratna in the year 1997.While delivering his speech at the Institute of Management, Shillong he got cardiac arrest and collapsed. He left us, making 27 July 2015 one of the saddest days in the history of India. To conclude, I would like to say even though he left us, he is still in our hearts. His golden words and miraculous deeds will always be remembered. In the end, I would like to thank you all for giving me this opportunity. Have a wonderful day."""
```

```python
text = re.sub(r'\[[0-9]*\]',' ',paragraph)
text = re.sub(r'\d',' ',text)
```

```python
text = re.sub(r'\s+',' ',paragraph)
```

```python
text = text.lower()
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
for i in range(len(sentences)):
  sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
model = Word2Vec(sentences,min_count=1)
```

```python
words = model.wv.key_to_index
```

```python
words
```

```python
vec = model.wv['management']
```

```python
similar = model.wv.most_similar('management')
```
